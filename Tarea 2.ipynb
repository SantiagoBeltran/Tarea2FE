```python
def factorial(n):
    a=1
    for i in range(1, n+1):
        a=a*i
    return a
factorial(5)
```

```python
n1=np.arange(10)
n1
```

```python
def proba(n1,n,s):
    proba=np.zeros(len(n1))
    for i in range(len(n1)):
        if n1[i]>=s and n1[i]<(n-s)/2:
            proba[i]=float(factorial(n))/float(factorial(n1[i]-s))*(1/3)**n
            proba[i]=proba[i]/float(factorial(n+s-2*n1[i])*factorial(n1[i]))
        else:
            proba[i]=0
    return proba
```

```python
proba(n1,10,0)
```

```python
plt.figure()
plt.scatter(n1,proba(n1,10,0))
```
