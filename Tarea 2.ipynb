```python
def factorial(n):
    a=1
    for i in range(1, n+1):
        a=a*i
    return a
factorial(5)
```

```python
def proba(n1,n,s):
    proba=np.zeros(len(n1))
    for i in range(len(n1)):
        if n1[i]>=s and n1[i]<(n+s)/2:
            proba[i]=float(factorial(n))/float(factorial(n1[i]-s))*(1/3)**n
            proba[i]=proba[i]/float(factorial(n+s-2*n1[i])*factorial(n1[i]))
        else:
            proba[i]=0
    return proba
```

```python
def promedio(n1,A):
    x11_prom=np.sum(n1[A!=0]*A[A!=0])
    return(x11_prom)
def sigma(n1,A):
    B=(n1[A!=0]**2)*A[A!=0]
    sigma11=np.sum(B)-(promedio(n1,A))**2
    return np.sqrt(sigma11)
```

```python
def combinatorio(a,b,c,d):
    k1=[b,c,d]
    k2=sorted(k1)
    n=1
    #print(k2[-1])
    for i in range(k2[-1]+1,a+1):
        #print(i)
        n=n*i
    #print("ya")
    #print(n)    
    n=n//(factorial(k2[-3]))
    n=n//(factorial(k2[-2]))
    return n
```

```python
def proba2(n1,n,s):
    proba=np.zeros(len(n1))
    for i in range(len(n1)):
        if n1[i]>=s and n1[i]<(n+s)/2:
            a=combinatorio(n,n1[i]-s,n+s-2*n1[i],n1[i])
            proba[i]=a
        else:
            proba[i]=0
    return proba
```
